```python
import numpy as np
```

```python
# let's solve Monty Hall Problem
```

```python
doors = [0, 1, 2]
prizes = ['goat', 'goat', 'car']
switch = True
choice = np.random.choice(doors)
wins = 0

# find the door that Monty will open
for i in range(3):
    if i !=choice and prizes[i] != 'car':
        open_door = i


if switch == True:
    for j in range(3):
        if j !=choice and j != open_door:
            choice_new = j

    if prizes[choice_new] == 'car':
        wins  = wins + 1

else:
    if prizes[choice] == 'car':
        wins  = wins + 1

print('door opened:', open_door, prizes[open_door], 'choice was:', choice, prizes[choice], 'switch:', choice_new)
```

```python
trials = 100000
wins = 0
doors = [0, 1, 2]
prizes = ['goat', 'goat', 'car']
switch = False

for t in range(trials):
    np.random.shuffle(prizes)
    choice = np.random.choice(doors)

    # find the door that Monty will open
    for i in range(3):
        if i !=choice and prizes[i] != 'car':
            open_door = i


    if switch == True:
        for j in range(3):
            if j !=choice and j != open_door:
                choice_new = j

        if prizes[choice_new] == 'car':
            wins  = wins + 1

    else:
        if prizes[choice] == 'car':
            wins  = wins + 1

print(wins/trials)
```
