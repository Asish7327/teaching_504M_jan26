```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# --- non-linear differential equation ---- 
def f(x):
    return x - x**3

#---- code
dt = 0.01
tmax = 10
t = np.arange(0, tmax + dt, dt)

initial_conditions = [0.2, 0.8, -0.4, 1.5, -2, 0]

for x0 in initial_conditions:
    x = np.zeros(len(t))
    x[0] = x0

    for i in range(len(t) - 1):
        x[i+1] = x[i] + dt * f(x[i])

    plt.plot(t, x, label=f"x(0) = {x0}")

plt.xlabel("t")
plt.ylabel("x(t)")
plt.title("Solutions of dx/dt = x - x^3")
plt.grid()
plt.legend()
plt.show()
```

```python
# --- Lorentz attractor - butterfly effect ----

# parameters
sigma = 10
rho = 28
beta = 8/3

def f(state):
    x, y, z = state
    dx = sigma * (y - x)
    dy = x * (rho - z) - y
    dz = x * y - beta * z
    return np.array([dx, dy, dz])

def rk4_step(state, dt):
    k1 = f(state)
    k2 = f(state + 0.5*dt*k1)
    k3 = f(state + 0.5*dt*k2)
    k4 = f(state + dt*k3)
    return state + (dt/6)*(k1 + 2*k2 + 2*k3 + k4)

# time
dt = 0.005
tmax = 40
steps = int(tmax/dt)

# initial condition
state = np.array([1.0, 1.0, 1.0])

xs, ys, zs = [], [], []

for _ in range(steps):
    xs.append(state[0])
    ys.append(state[1])
    zs.append(state[2])
    state = rk4_step(state, dt)

# plot
fig = plt.figure(figsize=(14,12))
ax = fig.add_subplot(projection='3d')

ax.plot(xs, ys, zs, lw=1, color = 'red')

# initial condition
state = np.array([1.00001, 1.0, 1.0])

xs, ys, zs = [], [], []

for _ in range(steps):
    xs.append(state[0])
    ys.append(state[1])
    zs.append(state[2])
    state = rk4_step(state, dt)

ax.plot(xs, ys, zs, lw=1, color = 'blue')

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title("Lorenz Attractor")

plt.show()
```

```python
# ============================================================
# Part (a): Plot square billiard
# ============================================================

def plot_square():
    L = 1.0
    plt.figure(figsize=(5,5))
    plt.plot([0, L, L, 0, 0], [0, 0, L, L, 0], 'k-')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Square billiard")
    plt.axis("equal")
    plt.grid()
    plt.show()

# plot_square()


# ============================================================
# Part (b): Plot square with circular obstacle
# ============================================================

def plot_square_circle():
    L = 1.0
    cx, cy = 0.5, 0.5
    R = 0.2

    theta = np.linspace(0, 2*np.pi, 400)

    plt.figure(figsize=(5,5))
    plt.plot([0, L, L, 0, 0], [0, 0, L, L, 0], 'k-')
    plt.plot(cx + R*np.cos(theta), cy + R*np.sin(theta), 'r-')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Square billiard with circular obstacle")
    plt.axis("equal")
    plt.grid()
    plt.show()

# plot_square_circle()


# ============================================================
# Part (c): One trajectory in square billiard
# ============================================================

def square_trajectory(x0=0.2, y0=0.3, vx=0.73, vy=0.41, dt=0.002, steps=5000):
    L = 1.0

    x, y = x0, y0
    xs, ys = [], []

    for _ in range(steps):
        x += vx * dt
        y += vy * dt

        # reflect from vertical walls
        if x <= 0:
            x = -x
            vx = -vx
        elif x >= L:
            x = 2*L - x
            vx = -vx

        # reflect from horizontal walls
        if y <= 0:
            y = -y
            vy = -vy
        elif y >= L:
            y = 2*L - y
            vy = -vy

        xs.append(x)
        ys.append(y)

    return np.array(xs), np.array(ys)

def plot_square_trajectory():
    xs, ys = square_trajectory()

    plt.figure(figsize=(5,5))
    plt.plot(xs, ys, lw=0.8)
    plt.plot([0,1,1,0,0], [0,0,1,1,0], 'k-')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("One trajectory in square billiard")
    plt.axis("equal")
    plt.grid()
    plt.show()

# plot_square_trajectory()


# ============================================================
# Part (d): Two nearby trajectories in square billiard
# ============================================================

def square_two_trajectories():
    dt = 0.0015
    steps = 4000

    x1s, y1s = square_trajectory(x0=0.2, y0=0.3, vx=0.73, vy=0.41, dt=dt, steps=steps)
    x2s, y2s = square_trajectory(x0=0.2001, y0=0.3, vx=0.73, vy=0.41, dt=dt, steps=steps)

    d = np.sqrt((x1s - x2s)**2 + (y1s - y2s)**2)
    t = np.arange(steps) * dt

    plt.figure(figsize=(5,5))
    #plt.plot(x1s, y1s, label="traj 1")
    #plt.plot(x2s, y2s, label="traj 2")
    plt.plot(x1s, y1s, lw=1.4, label="traj 1", color = 'blue')
    plt.plot(x2s, y2s, lw=1.2, label="traj 2", color = 'red', ls ='--')
    plt.plot([0,1,1,0,0], [0,0,1,1,0], 'k-')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Two nearby trajectories in square billiard")
    plt.axis("equal")
    plt.legend()
    plt.grid()
    plt.show()

    plt.figure(figsize=(7,4))
    plt.plot(t, d)
    plt.xlabel("time")
    plt.ylabel("distance")
    plt.title("Separation in square billiard")
    plt.grid()
    plt.show()

    return t, d

# t_square, d_square = square_two_trajectories()


# ============================================================
# Part (e): One trajectory in square with circular obstacle
# ============================================================

def sinai_step(x, y, vx, vy, L=1.0, cx=0.5, cy=0.5, R=0.2, dt=0.0015):
    x_new = x + vx * dt
    y_new = y + vy * dt

    # reflect from square walls
    if x_new <= 0:
        x_new = -x_new
        vx = -vx
    elif x_new >= L:
        x_new = 2*L - x_new
        vx = -vx

    if y_new <= 0:
        y_new = -y_new
        vy = -vy
    elif y_new >= L:
        y_new = 2*L - y_new
        vy = -vy

    # circle collision
    dist2 = (x_new - cx)**2 + (y_new - cy)**2
    if dist2 <= R**2:
        nx = x_new - cx
        ny = y_new - cy
        norm = np.sqrt(nx**2 + ny**2)
        nx = nx/ norm
        ny = ny/ norm

        dot = vx*nx + vy*ny
        vx = vx - 2*dot*nx
        vy = vy - 2*dot*ny

        # shift slightly outside circle
        x_new = cx + (R + 1e-4)*nx
        y_new = cy + (R + 1e-4)*ny

    return x_new, y_new, vx, vy

def sinai_trajectory(x0=0.2, y0=0.25, vx=0.83, vy=0.57, dt=0.0015, steps=12000):
    L = 1.0
    cx, cy = 0.5, 0.5
    R = 0.2

    # normalize velocity
    norm = np.sqrt(vx**2 + vy**2)
    vx /= norm
    vy /= norm

    x, y = x0, y0
    xs, ys = [], []

    for _ in range(steps):
        x, y, vx, vy = sinai_step(x, y, vx, vy, L=L, cx=cx, cy=cy, R=R, dt=dt)
        xs.append(x)
        ys.append(y)

    return np.array(xs), np.array(ys)

def plot_sinai_trajectory():
    xs, ys = sinai_trajectory()

    theta = np.linspace(0, 2*np.pi, 400)

    plt.figure(figsize=(5,5))
    plt.plot(xs, ys, lw=1)
    plt.plot([0,1,1,0,0], [0,0,1,1,0], 'k-')
    plt.plot(0.5 + 0.2*np.cos(theta), 0.5 + 0.2*np.sin(theta), 'r-')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("One trajectory in square billiard with circle")
    plt.axis("equal")
    plt.grid()
    plt.show()

# plot_sinai_trajectory()


# ============================================================
# Part (f): Two nearby trajectories in Sinai billiard
# ============================================================

def sinai_two_trajectories():
    dt = 0.0015
    steps = 15000

    x1s, y1s = sinai_trajectory(x0=0.2, y0=0.25, vx=0.83, vy=0.57, dt=dt, steps=steps)
    x2s, y2s = sinai_trajectory(x0=0.2001, y0=0.25, vx=0.83, vy=0.57, dt=dt, steps=steps)

    d = np.sqrt((x1s - x2s)**2 + (y1s - y2s)**2)
    t = np.arange(steps) * dt

    theta = np.linspace(0, 2*np.pi, 400)

    plt.figure(figsize=(5,5))
    plt.plot(x1s, y1s, lw=1.4, label="traj 1", color = 'blue')
    plt.plot(x2s, y2s, lw=1.2, label="traj 2", color = 'red', ls ='--')
    plt.plot([0,1,1,0,0], [0,0,1,1,0], 'k-')
    plt.plot(0.5 + 0.2*np.cos(theta), 0.5 + 0.2*np.sin(theta), 'g-')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Two nearby trajectories in square billiard with circle")
    plt.axis("equal")
    plt.legend()
    plt.grid()
    plt.show()

    plt.figure(figsize=(7,4))
    plt.plot(t, d)
    plt.xlabel("time")
    plt.ylabel("distance")
    plt.title("Separation in billiard with circle")
    plt.grid()
    plt.show()

    return t, d

# t_sinai, d_sinai = sinai_two_trajectories()


# ============================================================
# Part (g): Simple Lyapunov exponent estimate
# ============================================================

def plot_lyapunov_from_distance(t, d):
    d0 = d[0]

    # avoid divide-by-zero
    lam = np.log((d + 1e-12) / d0) / (t + 1e-12)

    plt.figure(figsize=(7,4))
    plt.plot(t[1:], lam[1:])
    plt.xlabel("time")
    plt.ylabel("lambda(t)")
    plt.title("Estimated Lyapunov exponent")
    plt.grid()
    plt.show()

    return lam

```

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(r, x):
    return r * x * (1 - x)

def plot_time_series(r, x0=0.5, n=250):
    x = x0
    xs = []

    for i in range(n):
        x = logistic_map(r, x)
        xs.append(x)

    plt.figure(figsize=(13,4))
    plt.plot(xs, marker='o', lw =0.5, markersize=4, markerfacecolor= 'g', color = 'b')
    plt.xlabel("n")
    plt.ylabel("x_n")
    plt.title(f"Logistic map time series (r = {r})")
    plt.grid()
    plt.show()

# Try different values
#plot_time_series(2.5)
#plot_time_series(3.2)
#plot_time_series(3.5)
plot_time_series(3.57)
#plot_time_series(3.82)
```

```python
# ---------------------------------
# Parameters
# ---------------------------------
r_values = np.linspace(2.5, 4.0, 2000)
n_steps = 1000
n_transient = 500

# ---------------------------------
# Storage for plotting
# ---------------------------------
R = []
X = []

# ---------------------------------
# Generate bifurcation diagram
# ---------------------------------
for r in r_values:
    x = 0.8  # initial condition
    
    for i in range(n_steps):
        x = r * x * (1 - x)
        
        # skip transient
        if i >= n_transient:
            R.append(r)
            X.append(x)

# ---------------------------------
# Plot
# ---------------------------------
plt.figure(figsize=(12,6))
plt.plot(R, X, ',k', alpha=0.25)  # tiny points
plt.xlabel('r')
plt.ylabel('x')
plt.title('Bifurcation Diagram of Logistic Map')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic(r, x):
    return r * x * (1 - x)

r_vals = np.linspace(2.5, 4, 2000)

lyap = []

for r in r_vals:
    x = 0.5
    s = 0
    for i in range(1000):
        x = logistic(r, x)
        if i > 100:
            s = s + np.log(abs(r * (1 - 2*x)))
    lyap.append(s / 900)

plt.plot(r_vals, lyap)
plt.axhline(0, color='k')
plt.xlabel("r")
plt.ylabel("Lyapunov exponent")
plt.show()
```
