```python
# illustration I
A = np.array([[2,1],
              [1,2]])

# compute eigenvectors
values, vectors = np.linalg.eig(A)

v1 = vectors[:,0]
v2 = vectors[:,1]

# create grid
x = np.linspace(-2,2,4)
y = np.linspace(-2,2,4)

plt.figure(figsize=(6,6))

for i in x:
    for j in y:

        v = np.array([i,j])
        v_new = A @ v

        # original vector (gray)
        plt.plot([0,v[0]],[0,v[1]],color='green')

        # transformed vector (blue)
        plt.plot([0,v_new[0]],[0,v_new[1]],color='blue', linestyle = '--')

# plot eigenvectors
#plt.quiver(0,0,v1[0],v1[1],scale=1,color='red',scale_units='xy', zorder = 200)
#plt.plot(0,0,v1[0],v1[1],scale=1,color='red',scale_units='xy')
#plt.quiver(0,0,v2[0],v2[1],scale=1,color='red',scale_units='xy', zorder = 200)

plt.xlim(-5,5)
plt.ylim(-5,5)
plt.grid()
plt.gca().set_aspect('equal')

plt.show()
```

```python
# illustration II
A = np.array([[2,1],
              [1,2]])

x = np.linspace(-2,2,21)

plt.figure(figsize=(6,6))

# original grid
for i in x:
    y = np.linspace(-2,2,50)
    pts = np.vstack([np.full_like(y,i), y])
    plt.plot(pts[0], pts[1], color='gray')

for j in x:
    x2 = np.linspace(-2,2,50)
    pts = np.vstack([x2, np.full_like(x2,j)])
    plt.plot(pts[0], pts[1], color='gray')

# transformed grid
for i in x:
    y = np.linspace(-2,2,50)
    pts = np.vstack([np.full_like(y,i), y])
    new = (A @ pts)
    plt.plot(new[0], new[1], color='blue')

for j in x:
    x2 = np.linspace(-2,2,50)
    pts = np.vstack([x2, np.full_like(x2,j)])
    new = (A @ pts)
    plt.plot(new[0], new[1], color='blue')

# eigenvectors
values, vectors = np.linalg.eig(A)
v1 = vectors[:,0]
v2 = vectors[:,1]

plt.quiver(0,0,v1[0],v1[1],color='red',scale=1,scale_units='xy', zorder =1000)
plt.quiver(0,0,v2[0],v2[1],color='red',scale=1,scale_units='xy', zorder = 1000)

plt.xlim(-6,6)
plt.ylim(-6,6)
plt.grid()
plt.gca().set_aspect('equal')

plt.show()
```

```python
# illustration II
A = np.array([[2,1],
              [1,2]])

x = np.linspace(-2,2,21)

plt.figure(figsize=(6,6))

# original grid
for i in x:
    y = np.linspace(-2,2,50)
    pts = np.vstack([np.full_like(y,i), y])
    plt.plot(pts[0], pts[1], color='gray')

for j in x:
    x2 = np.linspace(-2,2,50)
    pts = np.vstack([x2, np.full_like(x2,j)])
    plt.plot(pts[0], pts[1], color='gray')

# transformed grid
for i in x:
    y = np.linspace(-2,2,50)
    pts = np.vstack([np.full_like(y,i), y])
    new = A@(A@(A @ pts))
    plt.plot(new[0], new[1], color='blue')

for j in x:
    x2 = np.linspace(-2,2,50)
    pts = np.vstack([x2, np.full_like(x2,j)])
    new = A@(A@(A @ pts))
    plt.plot(new[0], new[1], color='blue')

# eigenvectors
values, vectors = np.linalg.eig(A)
v1 = vectors[:,0]
v2 = vectors[:,1]

plt.quiver(0,0,v1[0],v1[1],color='red',scale=1,scale_units='xy', zorder =1000)
plt.quiver(0,0,v2[0],v2[1],color='red',scale=1,scale_units='xy', zorder = 1000)

plt.xlim(-60,60)
plt.ylim(-60,60)
plt.grid()
plt.gca().set_aspect('equal')

plt.show()
```
