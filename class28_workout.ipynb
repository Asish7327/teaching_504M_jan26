```python
A = np.array([[2, 1, 3], [1, 2, 4], [-1, 3, -2]])
eigvalue, eigvector = np.linalg.eig(A)
print(eigvalue)
```

```python
#A = np.array([[2, 1], [1, 2]])
max_iter = 35

u = np.array([[7], [11], [17]])

for i in range(max_iter):
    v = A@u
    norm = np.sqrt(np.sum(v**2))
    u = v/norm

    lam = u.T@A@u/(u.T@u)

    print('eigenvalue = ', lam[0,0], 'eigenvector=', u[0,0], u[1,0], u[2,0])
```
