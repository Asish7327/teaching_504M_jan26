```python
n = 60 # number of students
birthdays = np.random.randint(366, size = n)
shared = False
for i in range(n):
    for j in range(i+1, n):
        #print(i, j)
        if birthdays[i] == birthdays[j]:
            shared = True 
            print('shared bday', i)
            
    #break
    if shared == True:
        break
```

```python
def is_bday_same(number_of_students):
    n= number_of_students
    birthdays = np.random.randint(366, size = n)
    shared = False
    for i in range(n):
        for j in range(i+1, n):
            if birthdays[i] == birthdays[j]:
                shared = True 
                #print('shared bday', i)          
        if shared == True:
            break
    return shared
```

```python
is_bday_same(number_of_students=10)
```

```python
class_size = 10
trials = 10000
count = 0
for t in range(trials):
    result = is_bday_same(number_of_students=class_size)
    if result == True:
        count = count + 1

p = count/trials
print(p)
```

```python
class_size = np.arange(2, 100)
p = []
for c in class_size:
    trials = 10000
    count = 0
    for t in range(trials):
        result = is_bday_same(number_of_students=c)
        if result == True:
            count = count + 1

    p.append(count/trials)
```
