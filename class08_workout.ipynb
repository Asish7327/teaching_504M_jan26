## Problem 1: Computing $e^x$ Using a Series Expansion

The exponential function can be written as a Taylor series:


$e^x = \sum_{n=0}^{\infty} \frac{x^n}{n!}$

### Tasks
1. Write a program to compute $e^x$ using the series expansion.
2. Use a loop to add terms of the series.
3. Stop the summation when the absolute value of the next term becomes smaller than $10^{-8}$.
4. Optional: Instead of step 3, usea  for loop to sum the first n terms
5. Evaluate $e^x$ for $x = 1$, $x = 5$, and $x = -2$.
6. Compare your result with Numpy `np.exp`.

```python
# using while loop
def exp_taylor(x, tol=1e-8):
    term = 1.0        # n = 0 term
    sum_exp = term
    n = 1

    while abs(term) > tol:
        term = term * x / n   # build next term from previous
        sum_exp = sum_exp + term
        n = n + 1

    return sum_exp, n
```

```python
exp_taylor(5)
```

```python
# using for loop
def exp_taylor_new(x, n_max = 100):
    term = 1.0        # n = 0 term
    sum_exp = term
    n = 1

    for n in range (1, n_max):
        term = term * x / n   # build next term from previous
        sum_exp = sum_exp + term
        # one can also write an if loop to check tolerance and break the for loop
        #if abs(term)<tol: # but define tol = 1e-8
        #    break
    

    return sum_exp
```

## Problem 2: Write a code to simulate Collatz Conjecture

* Collatz Conjecture:\
Take any positive integer $n$.\
If $n$ is even, divide it by $2$; if $n$ is odd, replace it by  $3n+1$.\
Repeat this process with the resulting number. \
The conjecture states that no matter which positive integer you start with, the sequence will eventually reach 1.

```python
def collatz(n, max_iter = 1000):
    print("Collatz simulation for number", n, end = " ---> ")
    for i in range (max_iter):
        if n == 1:
            print(int(n), ' STOP')
            break
        else:
            if n%2 == 0:
                n = n/2
            else:
                n = 3*n + 1
            print(int(n), end =" ---> ")
```

```python
collatz(13)
```

## Problem3 : Prime Numbers and Algorithmic Testing


### Tasks
1. Write a program to check whether a given integer $N$ is a prime number.
2. Use a loop to test divisibility of $N$ by integers starting from 2.
3. Stop checking as soon as a divisor is found.
4. Test your code for the following values:
   $
   N = 29,\; 51,\; 97,\; 121
   $
5. Extend your program to print **all prime numbers between 1 and 500**.
6. Count and print the total number of primes in this range.

```python
def is_prime(n):
    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True
```

```python
test_numbers = [29, 51, 97, 121]

for n in test_numbers:
    if is_prime(n):
        print(n, "is prime")
    else:
        print(n, "is not prime")
```

```python
primes = []

for n in range(2, 501):
    if is_prime(n):
        primes.append(n)

print("Prime numbers between 1 and 500:")
print(primes)
print("Total number of primes:", len(primes))
```

## NOT FOR THIS CLASS:  Problem 4: Free Fall with Air Drag (Euler Method)

A particle of mass $m$ falls vertically under gravity while experiencing linear air resistance.  
The equation of motion is

$
m\frac{dv}{dt} = mg - kv
$

where:
- $g = 9.8~\mathrm{m\,s^{-2}}$
- $m = 1~\mathrm{kg}$
- $k = 0.5~\mathrm{kg\,s^{-1}}$

The position $y(t)$ evolves according to

$
\frac{dy}{dt} = v
$

Initial conditions:
$
y(0) = 0, \quad v(0) = 0
$

---

### Euler Method

To solve these equations numerically, use **Euler’s method**, which approximates time derivatives using a finite time step $dt$.

At time $t_n$, the velocity and position are updated as:

$
v_{n+1} = v_n + a_n\,dt
$

$
y_{n+1} = y_n + v_n\,dt
$

where the acceleration is

$
a_n = g - \frac{k}{m}v_n
$

---
### Tasks
1. Choose a time step $dt = 0.01~\mathrm{s}$.
2. Use a loop to evolve the system from $t = 0$ to $t = 10~\mathrm{s}$.
3. At each step:
   - compute the acceleration,
   - update the velocity,
   - update the position.
4. Stop the simulation early if
   \[
   |v_{n+1} - v_n| < 10^{-4}
   \]
   indicating that terminal velocity has been reached.
5. Store time, velocity, and position in separate lists.
6. Plot:
   - velocity vs time,
   - position vs time.

```python
import matplotlib.pyplot as plt

g = 9.8
k = 0.5
m = 1.0
dt = 0.01
t_max = 10.0

v = 0.0
y = 0.0
t = 0.0

time = []
vel = []
pos = []

for i in range(int(t_max / dt)):
    time.append(t)
    vel.append(v)
    pos.append(y)

    a = g - (k/m) * v
    v_new = v + a * dt
    y = y + v * dt

    if abs(v_new - v) < 1e-4:
        v = v_new
        break

    v = v_new
    t += dt

plt.figure()
plt.plot(time, vel)
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.title("Velocity vs Time")

plt.figure()
plt.plot(time, pos)
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.title("Position vs Time")

plt.show()
```
