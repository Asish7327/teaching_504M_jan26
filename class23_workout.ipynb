```python
# plots for rotations
```

```python
theta = np.pi*90/180 # angle
A = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta), np.cos(theta)]])
v = np.array([[2], [2]])
```

```python
plt.plot([0, v[0,0]], [0, v[1, 0]] , marker = '.', color ='red')
plt.plot([0, (A@v)[0,0]], [0, (A@v)[1, 0]] , marker = '.', color ='blue', linestyle = '--')
plt.axhline(0, linestyle = ':')
plt.axvline(0, linestyle = ':')
```

```python
theta = np.pi*45/180 # angle
A = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta), np.cos(theta)]])

x = np.arange(-5, 5.1, 0.5)
y = np.arange(-5, 5.1, 0.5)

for i in range(len(x)):
    for j in range(len(y)):
        v = np.array([[x[i]], 
                      [y[j]]])
        plt.plot([0, v[0,0]], [0, v[1, 0]] , marker = '.', color ='red', alpha = 0.8)
        plt.plot([0, (A@v)[0,0]], [0, (A@v)[1, 0]] , marker = '.', color ='blue', linestyle = '--', zorder= 200, alpha =0.8)

plt.axhline(0, linestyle = ':')
plt.axvline(0, linestyle = ':')
plt.gca().set_aspect('equal')
```
