## Loops

* Three loops are the most important if, for and while
* the "while" loop is just a combination of if and for loops

#### if loop

```python
# just one if loop
a = 20
b = 30
if a is not b:
    print('a and b are different')
```

```python
# if and else loop
a=20
b=30
if a > b:
    print('right')
else:
    print('wrong')
```

```python
# more if conditions
a=40
b=20
if a > b:
    print('a > b')
elif a < b:
    print('a < b')
else:
    print('a=b')
```

```python
# you can also have nested loops
a=10
b=10
if a > b:
    
    if a-b == 1:
        print('a is b+1')
    else:
        print ('a > b but a is not b+1')

elif a<b:
    print('a less than or equal to b')
else:
    print('a = b')
```

### for loop

```python
# simple for loop and example of range
for i in range(6):
    print(i)
```

```python
range?
```

```python
# let's use a list in the for loop
my_list = [2, 4, 10, 6, 8]
print(my_list)
```

```python
# lets modify it
my_list.insert(1, 3)
print(my_list)
```

```python
help(my_list.append)
```

```python
# let's use two lists with the same length
my_list_new = [0.2, 0.3, 0.4, 1, 0.6, 0.8]
```

```python
# for loop to add these two lists by element-wise 
for l, m in zip(my_list, my_list_new):
    # see how I am using the zip function
    print('element in my_list', l, 'element in my_list_new', m, 'adding those:', l+m)
```

### while loop

```python
k = 1
while k <5:
    print(k)
    k = k + 1 # the stop condition
```
