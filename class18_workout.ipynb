```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5.1, 0.5)
```

```python
y = 3*x
```

```python
plt.scatter(x, y, facecolor = 'cyan', edgecolors='k', label = 'data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
y = 3 * x
np.random.seed(42)
#noise = np.random.uniform(-3, 3, len(y))
noise = np.random.normal(0, 3, len(y))
y = y + noise
```

```python
plt.scatter(x, y, facecolor = 'cyan', edgecolors='k', label = 'data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
slopes = np.arange(0, 6, 0.001)
```

```python
lsq = []
for m in slopes:
    f = np.sum((m*x-y)**2)
    lsq.append(f)
    
```

```python
plt.plot(slopes, lsq)
plt.xlabel('slope m')
plt.ylabel('least square')
```

```python
idx = np.argmin(lsq)
result = slopes[idx]
print('final m: ', result)
```
