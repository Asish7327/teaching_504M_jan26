```python
p_fair = 0.5
p_biased = 0.99

prior = 0.5  # part A
#prior = 0.1  # part B

experiments = 500000

max_k = 20

count_k = np.zeros(max_k+1)
count_k_and_biased =np.zeros(max_k+1)

for i in range(experiments):

    if np.random.rand() < prior:
        p = p_biased
        is_biased = True
    else:
        p = p_fair
        is_biased = False

    count_heads = 0
    while np.random.rand() <= p:
        count_heads = count_heads + 1
        if count_heads >= max_k:
            break

    count_k[count_heads] =  count_k[count_heads] + 1
    if is_biased:
        count_k_and_biased[count_heads] += 1
    
```

```python
# plotting
plt.plot(np.arange(max_k+1), count_k_and_biased/count_k, marker = '.', linestyle = '--', color = 'red', markersize = 10)
plt.xla
```
