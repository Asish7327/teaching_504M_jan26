```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    return y


def Euler(xf, h, x0 =0, y0 =1):

    x = np.arange(x0, xf+h, h)
    n = len(x)
    y = np.zeros(n)
    y[0] = y0
    x[0] = x0
    
    for i in range(n-1):
        y[i+1] = y[i] + h* f(x[i], y[i])

    return x, y


def RK4(xf, h, x0 =0, y0 =1):

    x = np.arange(x0, xf+h, h)
    n = len(x)
    y = np.zeros(n)
    y[0] = y0
    x[0] = x0
    
    for i in range(n-1):
        k1= f(x[i], y[i])
        k2= f(x[i]+h/2, y[i]+k1*h/2)
        k3= f(x[i]+h/2, y[i]+k2*h/2)
        k4= f(x[i]+h, y[i]+k3*h)
        y[i+1] = y[i] + (k1+2*k2+2*k3+k4)*h/6

    return x, y
```

```python
x, y = Euler(xf = 3, h=0.2)
plt.plot(x, y, marker= '.', color = 'b', linestyle = '--', label = 'Euler')
x, y = RK4(xf = 3, h=0.2)
plt.plot(x, y, marker= '.', color = 'r', linestyle = '--', label = 'Euler')
plt.plot(x, np.exp(x), color = 'k', label = 'exact' , alpha= 0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```
