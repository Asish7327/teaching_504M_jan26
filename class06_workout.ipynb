### Problem 5: Write code to simulate Collatz Conjecture

* Collatz Conjecture:\
Take any positive integer $n$.\
If $n$ is even, divide it by $2$; if $n$ is odd, replace it by  $3n+1$.\
Repeat this process with the resulting number. \
The conjecture states that no matter which positive integer you start with, the sequence will eventually reach 1.
