```python
def random_walk(Nsteps):
    x = 0
    for step in range(Nsteps):
        if np.random.random() < 0.5:
            x = x +1
        else:
            x = x -1
    return x
```

```python
trials = 10000
Nsteps = 27
p = []
for t in range(trials):
    xfinal = random_walk(Nsteps)
    p.append(xfinal)

```

```python
p = np.array(p)
np.mean(p**2)
```

```python
Nsteps = 100
x = 0
pos = [x]
for step in range(Nsteps):
    if np.random.random() < 0.5:
        x = x +1
    else:
        x = x -1
    pos.append(x)

steps = np.arange(Nsteps+1)
plt.plot(steps, pos)
plt.xlabel('steps')
plt.ylabel('x')
```

```python
Nsteps = 50
trials = 2000

x2 = []
for n in range (1, Nsteps):
    x = []
    for t in range(trials):
        x.append(random_walk(n))

    x2.append(np.mean(np.array(x)**2))
    #print(x, np.mean(np.array(x)**2))
    
steps = np.arange(1, Nsteps)  
plt.plot(steps, x2)
plt.grid()
plt.xlabel('steps')
plt.ylabel(r'x$^2$')
```
