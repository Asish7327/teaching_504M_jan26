### Solve following integral 
$\int_{a}^{b}{x^2}dx$ \
for interval (a, b) = (0,1)

```python
def f(x):
    return x**2
```

```python
a = 0 
b = 1
N = 1000 # steps
x = np.linspace(a, b, N)
```

### Reimann Sum 

```python
dx = x[1] - x[0]
I = dx * np.sum(f(x))
print('I = ', I)
```

### Trapazoidal Rule

```python
I = dx * (np.sum(f(x)) - 0.5*f(x[0]) - 0.5*f(x[-1]))
print('I = ', I)
```

### Monte Carlo integration

```python
N = 10000
x = np.random.uniform(a, b, N)
I =  (b-a)*np.mean(f(x))
print(I)
```

### Linear interpolation -- find adjucent points

```python
x = np.array([3, 2, 1, 4, 8, 9, 10, 5, 6])
y = f(x)
```

```python
# Note that the arrays X and Y are not sorted, so let's sort them first
idx = np.argsort(x) # gives indices of sorted array -- check by printing
```

```python
x_sorted = x[idx]
y_sorted = y[idx]
```

```python
# interpolte for value x0
x0 = 7
```

```python
# find two adjacent points
for i in range(len(x_sorted)-1):
    if x_sorted[i] <= x0 and x_sorted[i+1] >= x0:
        left = i
        right = i +1
```

```python
print(x_sorted[left], x_sorted[right] )
print(y_sorted[left], y_sorted[right] )
```
