```python
x= np.arange(-5, 5.1, 0.1)
```

```python
#x
```

```python
y= np.e**(-x**2)
```

```python
plt.plot(x, y)
plt.grid()
```

```python
plt.plot(x, y, marker= '.')
plt.grid()
```

```python
# how to save figure? and also add the legends 
plt.plot(x,y, 'o', c='red', markersize=6, markerfacecolor='cyan', alpha= 0.9, linestyle = '--', linewidth = 2, label = 'Gaussian')
plt.xlabel('x')
plt.ylabel(r'$e^{-x^{2}}$')
plt.legend()
plt.grid()
plt.savefig("class_example.pdf")
```
