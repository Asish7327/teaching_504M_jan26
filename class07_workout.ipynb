### 3. Write codes like cook-books. 

```python
def addition_of_elements(your_list):
    """
    arg:
    your_list = list or numpy array
    """
    if type(your_list) != list:
        print('your_list is not a list')
        return
    length_of_list = len(your_list)
    # initializing the sum
    total = 0
    for i in range(length_of_list):
        # summing elements
        total = total + your_list[i]
    return total
```

```python
addition_of_elements(2)
```

```python
v = 10 #m/s
theta = 45 # degrees
theta_rad = theta*np.pi/180
```

```python
vx = v* np.sin(theta_rad)
vy = v* np.cos(theta_rad)
t = np.arange(0, 2*vy/9.8, 0.001)
x = vx*t
y = vy*t - 0.5*9.8*t**2
print(x[-1])
```

```python
plt.plot(x, y)
```

```python
v2 = v/np.sqrt(2)
vx = v2* np.sin(theta_rad)
vy = v2* np.cos(theta_rad)
t = np.arange(0, 2*vy/9.8, 0.001)
x2 = vx*t 
y2 = vy*t - 0.5*9.8*t**2
```

```python
plt.plot(x, y)
plt.plot(x2 + x[-1], y2)
```
