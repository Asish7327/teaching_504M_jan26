### Problem 1: Write a function to print a multiplication table for number n    

```python
# soltuion 
def multiplication_table(n):
    for i in range(10):
        print(n*(i+1))
    return
```

```python
multiplication_table(9)
```

### Problem 2: Write a function to find the sum of n natural numbers.    

```python
# given n, find the sum of 1+2+...+n-1+n
def sum_of_n_numbers(n):
    sum = 0
    for i in range(1, n+1):
        sum = sum + i
    return sum
```

```python
n=100
print(sum_of_n_numbers(n=n))
```

### Problem 3: Write a custom function to find a maximum element and its index in a list  

```python
def find_max(l):
    """
    arg:
    l = list
    returns
    maximum element in the list
    """

    # find the length of the list
    n = len(l)
    init = 0
    for i in range(n-1):
        max_ind = init
        if l[i] >= l[i+1]:
            l[i+1] = l[i]
            max_val = l[i]
        else:
            max_val = l[i+1]
            max_ind = i+1
            init  = i+1

    print('max value', max_val)
    print('max index', max_ind)
    return 
```

```python
find_max([1, 3, 9, 2, 11, 3.5, 35, 0.5])
```

### Problem 4: Plot height of a mass m thrown from the ground with initial velocity with time 

```python
def vertical_distance(t, u, g=-9.8):
    """
    args:
    u in m/s initial vertical velocity component
    t array
    returns:
    y or vertical distance values for time array
    """
    return u*t + 0.5*g*t**2
```

```python
import numpy as np
t = np.arange(0, 2, 0.05)
u = 10.
y = vertical_distance(t, u)
```

```python
import matplotlib.pyplot as plt
plt.plot(t, y, marker='.', markersize = 14, linestyle ='--', c= 'red', markerfacecolor = 'cyan', linewidth = 1.5)
plt.grid()
plt.xlabel('time')
plt.ylabel('y')
```
