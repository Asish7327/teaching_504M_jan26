### Problem 1: Write a function to print a multiplication table for number n    

```python
def multiplication_table(n):
    for i in range(1,11):
        print(n*i)
    return
```

### Problem 2: Write a function to find the sum of n natural numbers.    

```python
def sum_of_n_numbers(n):
    s=0
    for i in range(n+1):
        s = s + i
    print(s)
        
    
```

```python
sum_of_n_numbers(100)
```

### Problem 3: Write a custom function to find a maximum element and its index in a list  

```python
l = [1, 4, 8, 2, 9, 45, 23]
```

```python
print(len(l))
```

```python
# this answer is better than the one in the notes
def max_element(l):
    n = len(l)
    max_number = l[0]
    max_index = 0
    for i in range(n-2):
        if l[i+1] > max_number:
            max_number = l[i+1]
            max_index = i+1
    print(max_number, 'max number')
    print(max_index, 'max_index')
            
```

```python
max_element([40,2,800, 20, 2.4])
```
