```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
## Heat equation in 1D

# ----------------------------
# Parameters
# ----------------------------
L = 1.0              # length of rod
Nx = 101             # number of spatial points
dx = L / (Nx - 1)

alpha = 0.01         # thermal diffusivity

dt = 0.0004          # time step
Nt = 1000            # number of time steps

x = np.linspace(0, L, Nx)

# Stability parameter for FTCS
r = alpha * dt / dx**2
print("r =", r)

# FTCS is stable only if r <= 0.5
if r > 0.5:
    print("Warning: scheme may be unstable!")

# ----------------------------
# Initial condition
# ----------------------------
# A Gaussian hot bump in the middle
u = np.exp(-200 * (x - 0.5)**2)

# Boundary conditions: fixed ends
u[0] = 0
u[-1] = 0

# Store solution for animation
solution = [u.copy()]

# ----------------------------
# Time evolution
# ----------------------------
for n in range(Nt):
    u_new = u.copy()

    for i in range(1, Nx - 1):
        u_new[i] = u[i] + r * (u[i+1] - 2*u[i] + u[i-1])

    # fixed boundary conditions
    u_new[0] = 0
    u_new[-1] = 0

    u = u_new
    solution.append(u.copy())

solution = np.array(solution)

# ----------------------------
# Animation
# ----------------------------
fig, ax = plt.subplots(figsize=(8, 4))
line, = ax.plot(x, solution[0], lw=2, color = 'red', marker = '.', markerfacecolor = 'g', markersize = 10)
ax.set_xlim(0, L)
ax.set_ylim(0, 1.1)
ax.set_xlabel("x")
ax.set_ylabel("Temperature u(x,t)")
ax.set_title("1D Heat Equation")

time_text = ax.text(0.02, 0.92, "", transform=ax.transAxes)

def update(frame):
    line.set_ydata(solution[frame])
    time_text.set_text(f"step = {frame}")
    return line, time_text

anim = FuncAnimation(fig, update, frames=len(solution), interval=70, blit=True)

plt.close(fig)
HTML(anim.to_jshtml())
```

```python
# 1D wave equation - initial velocity = 0

# ============================================================
# 1D Wave Equation:
#     d^2u/dt^2 = c^2 d^2u/dx^2
# ============================================================

# ----------------------------
# Parameters
# ----------------------------
L = 1.0                 # Length of string
Nx = 200                # Number of spatial points
dx = L / (Nx - 1)

c = 1.0                 # Wave speed
dt = 0.004              # Time step
Nt = 400                # Number of time steps

x = np.linspace(0, L, Nx)

# Courant number
r = c * dt / dx
print("Courant number r =", r)

# Stability check
if r > 1:
    print("Warning: Scheme may be unstable because r > 1")

# ----------------------------
# Arrays
# ----------------------------
u = np.zeros((Nt, Nx))

# ----------------------------
# Initial condition: Gaussian pulse
# ----------------------------
u[0, :] = np.exp(-1000 * (x - 0.5)**2)

# Fixed boundary conditions
u[0, 0] = 0
u[0, -1] = 0

# ----------------------------
# Initial velocity = 0
# Compute first time step using Taylor expansion
# ----------------------------
for i in range(1, Nx - 1):
    u[1, i] = (u[0, i]
               + 0.5 * r**2 * (u[0, i+1] - 2*u[0, i] + u[0, i-1]))

u[1, 0] = 0
u[1, -1] = 0

# ----------------------------
# Time evolution
# ----------------------------
for n in range(1, Nt - 1):
    for i in range(1, Nx - 1):
        u[n+1, i] = (2*u[n, i] - u[n-1, i]
                     + r**2 * (u[n, i+1] - 2*u[n, i] + u[n, i-1]))

    # Fixed boundaries
    u[n+1, 0] = 0
    u[n+1, -1] = 0

# ============================================================
# Animation
# ============================================================
fig, ax = plt.subplots(figsize=(8, 4))
line, = ax.plot(x, u[0], lw=2)
ax.set_xlim(0, L)
ax.set_ylim(-1.1, 1.1)
ax.set_xlabel("x")
ax.set_ylabel("u(x,t)")
ax.set_title("1D Wave Equation")

time_text = ax.text(0.02, 0.90, "", transform=ax.transAxes)

def update(frame):
    line.set_ydata(u[frame])
    time_text.set_text(f"Time step = {frame}")
    return line, time_text

ani = FuncAnimation(fig, update, frames=Nt, interval=20, blit=True)

plt.close(fig)  # Prevents duplicate static plot in Jupyter
HTML(ani.to_jshtml())
```

```python
# 2D wave equation

import cmocean
# comment above if you don't have this

# ============================================================
# 2D Wave Equation
# d^2u/dt^2 = c^2 (d^2u/dx^2 + d^2u/dy^2)
# ============================================================

# ----------------------------
# Parameters
# ----------------------------
Lx, Ly = 1.0, 1.0
Nx, Ny = 100, 100

dx = Lx / (Nx - 1)
dy = Ly / (Ny - 1)

c = 1.0
dt = 0.004
Nt = 400

x = np.linspace(0, Lx, Nx)
y = np.linspace(0, Ly, Ny)
X, Y = np.meshgrid(x, y)

# Stability check (rough condition for equal dx=dy)
r = c * dt / dx
print("Courant-like number r =", r)
if r >= 1 / np.sqrt(2):
    print("Warning: dt may be too large for stability in 2D.")

# ----------------------------
# Arrays
# ----------------------------
u_prev = np.zeros((Ny, Nx))   # u at time n-1
u_now  = np.zeros((Ny, Nx))   # u at time n
u_next = np.zeros((Ny, Nx))   # u at time n+1

# ----------------------------
# Initial condition: Gaussian bump in center
# ----------------------------
u_now = np.exp(-1000 * ((X - 0.5)**2 + (Y - 0.5)**2))

# Fixed boundaries
u_now[0, :] = 0
u_now[-1, :] = 0
u_now[:, 0] = 0
u_now[:, -1] = 0

# Initial velocity = 0
u_prev[:, :] = u_now[:, :]

# To store frames for animation
frames = []

# Store every few time steps for smoother/faster animation
skip = 2

# ----------------------------
# Time evolution
# ----------------------------
for n in range(Nt):
    # Interior update
    u_next[1:-1, 1:-1] = (
        2 * u_now[1:-1, 1:-1] - u_prev[1:-1, 1:-1]
        + (c * dt / dx)**2 * (
            u_now[1:-1, 2:] - 2*u_now[1:-1, 1:-1] + u_now[1:-1, :-2]
        )
        + (c * dt / dy)**2 * (
            u_now[2:, 1:-1] - 2*u_now[1:-1, 1:-1] + u_now[:-2, 1:-1]
        )
    )

    # Fixed boundaries
    u_next[0, :] = 0
    u_next[-1, :] = 0
    u_next[:, 0] = 0
    u_next[:, -1] = 0

    if n % skip == 0:
        frames.append(u_now.copy())

    # Shift time levels
    u_prev, u_now, u_next = u_now, u_next, u_prev

# ============================================================
# Animation
# ============================================================
fig, ax = plt.subplots(figsize=(6, 5))

img = ax.imshow(
    frames[0],
    extent=[0, Lx, 0, Ly],
    origin='lower',
    cmap=cmocean.cm.dense,
    #cmap = 'grey',
    vmin=-0.2, vmax=0.2,
    animated=True
)

ax.set_title("2D Wave Equation: Water-like Ripples", pad = 20)
ax.set_xlabel("x")
ax.set_ylabel("y")

time_text = ax.text(
    0.02, 1.02, "", transform=ax.transAxes
)

def update(frame):
    img.set_array(frames[frame])
    time_text.set_text(f"Frame = {frame}")
    return img, time_text

ani = FuncAnimation(fig, update, frames=len(frames), interval=40, blit=True)

plt.close(fig)
HTML(ani.to_jshtml())
```

```python
# 1D wave equation with velocity to the right 

# ============================================================
# 1D Wave Equation:
#     d^2u/dt^2 = c^2 d^2u/dx^2
# Right-moving Gaussian pulse
# ============================================================

# Parameters
L = 1.0
Nx = 300
dx = L / (Nx - 1)

c = 1.0
dt = 0.002
Nt = 600

x = np.linspace(0, L, Nx)
r = c * dt / dx
print("Courant number r =", r)

if r > 1:
    print("Warning: unstable because r > 1")

# Arrays
u = np.zeros((Nt, Nx))

# ------------------------------------------------
# Initial shape f(x): Gaussian
# ------------------------------------------------
f = np.exp(-400 * (x - 0.3)**2)

# Its derivative df/dx
dfdx = -800 * (x - 0.3) * np.exp(-400 * (x - 0.3)**2)

# Initial displacement
u[0, :] = f

# Initial velocity for right-moving wave:
# v(x,0) = -c * df/dx
v0 = -c * dfdx

# Fixed boundaries
u[0, 0] = 0
u[0, -1] = 0

# ------------------------------------------------
# First time step using Taylor expansion
# u_i^1 = u_i^0 + dt*v_i^0 + 0.5*r^2*(u_{i+1}^0 - 2u_i^0 + u_{i-1}^0)
# ------------------------------------------------
for i in range(1, Nx - 1):
    u[1, i] = (u[0, i]
               + dt * v0[i]
               + 0.5 * r**2 * (u[0, i+1] - 2*u[0, i] + u[0, i-1]))

u[1, 0] = 0
u[1, -1] = 0

# ------------------------------------------------
# Time evolution
# ------------------------------------------------
for n in range(1, Nt - 1):
    for i in range(1, Nx - 1):
        u[n+1, i] = (2*u[n, i] - u[n-1, i]
                     + r**2 * (u[n, i+1] - 2*u[n, i] + u[n, i-1]))

    u[n+1, 0] = 0
    u[n+1, -1] = 0

# ============================================================
# Animation
# ============================================================
fig, ax = plt.subplots(figsize=(8, 4))
line, = ax.plot(x, u[0], lw=2)
ax.set_xlim(0, L)
ax.set_ylim(-1.1, 1.1)
ax.set_xlabel("x")
ax.set_ylabel("u(x,t)")
ax.set_title("1D Wave Equation: Right-Moving Gaussian Pulse")
time_text = ax.text(0.02, 0.90, "", transform=ax.transAxes)

def update(frame):
    line.set_ydata(u[frame])
    time_text.set_text(f"Time step = {frame}")
    return line, time_text

ani = FuncAnimation(fig, update, frames=Nt, interval=20, blit=True)

plt.close(fig)
HTML(ani.to_jshtml())
```
